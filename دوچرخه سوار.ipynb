```python
#حل معادله حرکت دورخه سوار با روش اویلر
import numpy as np
import matplotlib.pyplot as plt

#شرایط اولیه
v_0=4          #سرعت اولیه
P=400          #توان
h=0.1          #گام زمانی
t_tot=200     #زمان کل
m=70           # جرم ادم و دوچرخه
c=0.5          #drg coefficient
d=1            #چگالی هوا
A=0.33         #مساحت

# اعداد گام ها
N=int(t_tot/h)

#لیست زمان و سرعت
t_list=[0]
t_air_list=[0]
v_list=[v_0]
v_air_list=[v_0]

# الگوریتم اویلر حالت بدون مقاوت هوا
for n in range(1, N):
   t_new= n*h
   v_new=v_list[-1]+(h*P/(m*(v_list[-1])))
   t_list.append(t_new)
   v_list.append(v_new)

#الگوریتم اویلر با مقاومت هوا 
for k in range(1, N):
   t_air_new = k*h
   v_air_new= v_air_list[-1] + (h*P/(m*(v_air_list[-1]))) +(-1*c*h*d*A*(v_air_list[-1]**2)/m)
   t_air_list.append(t_air_new)
   v_air_list.append(v_air_new)


#تبدیل لیست ها به آرایه 
t_euler = np.array(t_list)
v_euler = np.array(v_list)
t_air_euler = np.array(t_air_list)
v_air_euler = np.array(v_air_list)


#رسم نمودار 
plt.plot(t_euler, v_euler, color ='red', label= 'Bicyclig without air resistance')
plt.plot(t_air_euler, v_air_euler, color ='blue', label= 'Bicycling with air resistance')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.legend()
plt.show()
```
